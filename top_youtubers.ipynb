**Data set cleaning**

```python
# Drop rows with missing values in crucial columns
df_hr = df_hr.drop(columns=['Rank', 'Channel Name', 'Subscribers'], axis=1)
```

```python
df_hr
```

```python
# Define numeric columns
numeric_cols = ['Average Views', 'Average Likes', 'Average Comments']
```

```python
# Convert numeric columns to numeric types
df_hr[numeric_cols] = df_hr[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Handle missing values in 'Category', 'Country', and 'Content Type'
# Fill missing values in 'Category' with a placeholder value "Unknown"
df_hr['Category'] = df_hr['Category'].fillna("Unknown")
```

```python
# Fill missing values in 'Country' with a placeholder value "Unknown"
df_hr['Country'] = df_hr['Country'].fillna("Unknown")
```

```python
# Fill missing values in 'Content Type' with a placeholder value "Unknown"
df_hr['Content Type'] = df_hr['Content Type'].fillna("Unknown")

df_hr
```

```python
# Inpute missing values with the mean for numeric columns
df_hr[['Average Views', 'Average Likes', 'Average Comments']] = df_hr[['Average Views', 'Average Likes', 'Average Comments']].apply(lambda x: x.fillna(x.mean()))
df_hr
```
